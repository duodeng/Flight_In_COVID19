```python
import os

directory = 'D:/Germany/data analysis/Ironhack/course/dataset/flight/after extract'

files = []
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        files.append(os.path.join(directory, filename))
       
```

```python
files
```

```python
import os
import pandas as pd
   
dataframes = {}
for file in files:
    df = pd.read_csv(file)
    columns_to_drop = ['number', 'icao24', 'registration', 'typecode', 'origin', 'destination', 'altitude_1', 'altitude_2']
    df.drop(columns_to_drop, axis=1, inplace=True)
    key = 'f' + file[-12:-4]
    dataframes[file] = df
```

```python
dataframes.keys()
```

```python
dataframes_s = {f'f{k[-12:-8]+ k[-8:-6]}': v for k, v in dataframes.items()}

print(dataframes_s)
```

```python
dataframes_s.keys()
```

```python
for key, df in dataframes_s.items():
    df['callsign'] = df['callsign'].str[:3]
```

```python
dataframes_s['f201911']
```

```python
pass_top10 = ['THY','AFR','BAW','ETH','UAE','QTR','DLH','AAL','UAL','KLM']
```

```python
cargo_top5 = ['CLX','FDX','GEC','GTI','UPS']
```

```python
filter_list = pass_top10 + cargo_top5

print(filter_list)
```

```python
for key, df in dataframes_s.items():
    df = df[df['callsign'].isin(filter_list)]
    dataframes_s[key] = df
```

```python
 dataframes_s['f201911'].info()
```

```python
flight_df = pd.concat(dataframes_s.values(), ignore_index=True)
```
