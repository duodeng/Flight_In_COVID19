```python
df_2020_01 = df_city_flight[df_city_flight["Month of Date"] == "2020-01-01"]
df_2022_10 = df_city_flight[df_city_flight["Month of Date"] == "2022-10-01"]

difference = df_2022_10.merge(df_2020_01, on="City", suffixes=("_2022_10", "_2020_01"))
difference["Increase Percentage"] = (difference["Number of Active Flights_2022_10"] - difference["Number of Active Flights_2020_01"]) / difference["Number of Active Flights_2020_01"]

df_city_flight = df_city_flight.merge(difference[["City", "Increase Percentage"]], on="City")
```
