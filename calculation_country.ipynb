```python
df_2020_01 = df_country_flight[df_country_flight["Month of Date"] == "2020-01-01"]
df_2022_10 = df_country_flight[df_country_flight["Month of Date"] == "2022-10-01"]

difference = df_2022_10.merge(df_2020_01, on="Country", suffixes=("_2022_10", "_2020_01"))
difference["Increase Percentage"] = (difference["Number of Active Flights_2022_10"] - difference["Number of Active Flights_2020_01"]) / difference["Number of Active Flights_2020_01"]

df_country_flight = df_country_flight.merge(difference[["Country", "Increase Percentage"]], on="Country")
```
