```python
import numpy as np
import pandas as pd
```

```python
covid_df = pd.read_csv('WHO-COVID-19-global-data.csv')
```

```python
covid_df
```

```python
# drop irrelevant columns 
columns_to_drop = ['Country_code', 'WHO_region', 'Cumulative_cases', 'Cumulative_deaths']
covid_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
# delete duplicated rows
covid_df.drop_duplicates(inplace = True)
```

```python
# reset index
covid_df.reset_index(drop = True, inplace = True)
```

```python
covid_df[['Date_reported']] = covid_df[['Date_reported']].apply(pd.to_datetime)
```

```python
covid_df.info()
```

```python
covid_df.describe()
```

```python
# set negative values in the columns 'New_cases' and 'New_deaths' to 0
covid_df[['New_cases', 'New_deaths']] = covid_df[['New_cases', 'New_deaths']].clip(lower=0)
```

```python
covid_df.describe()
```

```python
covid_df.to_csv("data/covid_cleaned.csv")
```
